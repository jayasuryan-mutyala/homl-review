```python
corpus = """Hello Welcome, to Krish Naik's NLP Tutorials.
Please do watch the entire course! to become an expert in NLP.
"""
```

```python
print(corpus)
```

```python
# Converting a paragraph to sentences

from nltk.tokenize import sent_tokenize

documents = sent_tokenize(corpus)
print(documents)
print(type(documents))
```

```python
for sentence in documents:
    print(sentence)
```

```python
# Converting a paragraph to words and sentence to words 

from nltk.tokenize import word_tokenize

words = word_tokenize(corpus)
words
```

```python
from nltk.tokenize import wordpunct_tokenize 

words = wordpunct_tokenize(corpus)
words
```

```python
from nltk.tokenize import TreebankWordTokenizer

# fullstop is not treated as a seperate word in the first few lines 
# only with respect to the last line it considers it as a seperate word

tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
