```python
import nltk 

sentence = "The Eiffel Tower was build from 1887 to 1889 by French emgineer Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
import nltk 

words = nltk.word_tokenize(sentence)
```

```python
tag_elements = nltk.pos_tag(words)
```

```python
nltk.ne_chunk(tag_elements).draw()
```
