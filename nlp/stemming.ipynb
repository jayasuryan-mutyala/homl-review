```python
words = ["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
from nltk.stem import PorterStemmer

stemming = PorterStemmer()
```

```python
for word in words:
    print(word +" ---> "+stemming.stem(word))
```

```python
from nltk.stem import RegexpStemmer

# stemming method that uses the regular expression rules to remove or 
# replace word ending
 
reg_stemmer = RegexpStemmer(regexp='ing$|s$|e$|able$',min=4)
# regexp will remove words ending with ing or s or e or able
```

```python
reg_stemmer.stem("running")
reg_stemmer.stem("beable")
```

```python
reg_stemmer.stem("beind")
```

```python
from nltk.stem import SnowballStemmer

snow_ball_stemmer = SnowballStemmer("english")
```

```python
for word in words:
    print(word + " ---> " + snow_ball_stemmer.stem(word))
```

```python
# comparing porter stemer and snowball stemmer 

stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
snow_ball_stemmer.stem("fairly"),snow_ball_stemmer.stem("sportingly")
```
